# Part 4: Prior and Posterior Distributions

```python
#full importing template
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
print("Enter the value of alpha: ")
alpha_prior = int(input())
print("Enter the value of beta: ")
beta_prior = int(input())
```

```python
print("Enter the likes of 1st survey: ")
likes_first = int(input())
print("Enter the total respondents of 1st survey")
total_first = int(input())

print("Enter the dislikes of 2nd survey: ")
dislikes_second = int(input())
print("Enter the total respondents of 2nd survey")
total_second = int(input())
```

```python
dislikes_first = total_first - likes_first
likes_second = total_second - dislikes_second
```

```python
alpha_first = alpha_prior + likes_first
beta_first = beta_prior + dislikes_first

alpha_second = alpha_first + likes_second
beta_second = beta_first + dislikes_second
```

```python
print(alpha_prior)
print(beta_prior)
print(alpha_first)
print(beta_first)
print(alpha_second)
print(beta_second)
```

```python
p_values = np.linspace(0, 1, 1000)   # generate an array of 1000 evenly spaced values between 0 and 1.
prior_pdf = beta.pdf(p_values, alpha_prior, beta_prior)
posterior_first_pdf = beta.pdf(p_values, alpha_first, beta_first)
posterior_second_pdf = beta.pdf(p_values, alpha_second, beta_second)
```

```python
#Plotting template
plt.plot(p_values, prior_pdf)
plt.title('Prior Distribution')
plt.xlabel('p')
plt.ylabel('Density')
```

```python
plt.plot(p_values, posterior_first_pdf)
plt.title('Posterior Distribution after First Survey')
plt.xlabel('p')
plt.ylabel('Density')
```

```python
plt.plot(p_values, posterior_second_pdf)
plt.title('Posterior Distribution after Second Survey')
plt.xlabel('p')
plt.ylabel('Density')
```
