# P7: Principal Component Analysis (PCA) on Gym Members Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import beta
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

## Data Import and Cleaning

```python
df = pd.read_csv("../dataset/gym_membership.csv")
df
```

```python
df_cleaned = df.drop(columns=["id","name_personal_trainer","birthday","avg_time_check_out","avg_time_check_in","days_per_week"])
```

```python
df_cleaned
```

```python
missing_values = df_cleaned.isnull().sum()
print(missing_values)
```

```python
df_cleaned = df_cleaned.drop(columns=["fav_group_lesson","fav_drink"])
df_cleaned
```

```python
df_cleaned.info()
```

```python
df_cleaned.describe()
```

## Data Scaling and Visualization

### Feature Encoding

```python
boolean_columns = ["attend_group_lesson","drink_abo","personal_training","uses_sauna"]
for column in boolean_columns:
    if column in df_cleaned.columns:
        df_cleaned[column] = df_cleaned[column].astype(int)
```

```python
df_cleaned.head()
```

```python
categorical_columns = ["gender","abonoment_type"]
gender_mapping = {'Male': 1, 'Female':0}
df_cleaned["gender"] = df_cleaned["gender"].map(gender_mapping)
df_cleaned.head()
```

```python
df_cleaned["abonoment_type"].unique()
```

```python
abonoment_type_mapping = {'Premium': 1, 'Standard': 0}
df_cleaned["abonoment_type"] = df_cleaned["abonoment_type"].map(abonoment_type_mapping)
df_cleaned.head()
```

### Data Scaling

```python
numerical_columns = ["Age","visit_per_week","avg_time_in_gym"]
scaler = StandardScaler()
df_scaled = df_cleaned.copy()
df_scaled[numerical_columns] = scaler.fit_transform(df_scaled[numerical_columns])
df_scaled.head()
```

```python
visualize_columns = ["visit_per_week","avg_time_in_gym","abonoment_type"]
sns.pairplot(df_scaled[visualize_columns])
plt.figure(figsize=(12, 8))
plt.show()
```

```python
correlation_matrix = df_scaled.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.4f')
plt.title('Correlation Heatmap')
plt.show()
```

## PCA Application

### Apply PCA

```python
pca = PCA()
pca.fit(df_scaled)
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
num_components_80 = np.argmax(cumulative_explained_variance >= 0.80) + 1
num_components_90 = np.argmax(cumulative_explained_variance >= 0.90) + 1
print(f"Number of components to cover 80% variance: {num_components_80}")
print(f"Number of components to cover 90% variance: {num_components_90}")
```

```python
plt.figure(figsize=(12, 8))
plt.plot(cumulative_explained_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance by PCA Components')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()
```

## Visualization of PCA Results

### 2D Scatter Plot

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df_scaled)
```

```python
principal_components
```

```python
pca_df = pd.DataFrame(data=principal_components, columns=["PC1","PC2"])
pca_df["abonoment_type"] = df_scaled["abonoment_type"]
```

```python
pca_df.head()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='PC1',y='PC2',hue='abonoment_type',data=pca_df)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

### Loadings Plot

```python
# Get the loadings (eigenvectors)
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

# Plot the loading vectors
for i, feature in enumerate(df_scaled.columns):
    plt.arrow(0, 0, loadings[i, 0], loadings[i, 1])
    plt.text(loadings[i, 0] * 1.15, loadings[i, 1] * 1.15, feature, color='g')

plt.show()
```

- "Visit_per_week" contributes strongly to PC1 and slightly to PC2, as its arrow is long and leans towards the positive horizontal axis.
- "Avg_time_in_gym" contributes strongly to PC1 but not much to PC2 (mostly horizontal).
- "Age" contributes very little to both PC1 and PC2, as its arrow is short and almost centered.
- Features like "gender" and "person_to_trainer_ratio" contribute minimally since their arrows are short and closer to the origin.
